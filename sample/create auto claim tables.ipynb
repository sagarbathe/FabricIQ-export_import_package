# Create AutoClaims Sample Tables

Creates all delta tables used by the AutoClaims sample ontology in the attached lakehouse.

**Tables created:** `Policyholder`, `Vehicle`, `Policy`, `Adjuster`, `Accident`, `Claim`, `Driver_Telemetry_Data`

```python
%%sql
DROP TABLE IF EXISTS Policyholder;
DROP TABLE IF EXISTS Vehicle;
DROP TABLE IF EXISTS Policy;
DROP TABLE IF EXISTS Adjuster;
DROP TABLE IF EXISTS Accident;
DROP TABLE IF EXISTS Claim;
DROP TABLE IF EXISTS Driver_Telemetry_Data;
```

```python
%%sql

CREATE TABLE IF NOT EXISTS Policyholder (
    id BIGINT,
    name STRING,
    address STRING,
    city STRING,
    state STRING,
    phone STRING,
    email STRING,
    date_of_birth DATE
);

CREATE TABLE IF NOT EXISTS Vehicle (
    vin STRING,
    make STRING,
    model STRING,
    year INT
);

CREATE TABLE IF NOT EXISTS Policy (
    policy_number STRING,
    policyholder_id BIGINT,
    vehicle_vin STRING,
    start_date DATE,
    end_date DATE,
    coverage_type STRING,
    premium DOUBLE
);

CREATE TABLE IF NOT EXISTS Adjuster (
    id BIGINT,
    name STRING,
    phone STRING,
    email STRING
);

CREATE TABLE IF NOT EXISTS Accident (
    accident_id BIGINT,
    policyholder_id BIGINT,
    vehicle_vin STRING,
    accident_date DATE,
    location STRING,
    accident_type STRING,
    severity STRING
);

CREATE TABLE IF NOT EXISTS Claim (
    claim_number STRING,
    policy_number STRING,
    accident_id BIGINT,
    date_filed DATE,
    status STRING,
    adjuster_id BIGINT,
    claim_amount DOUBLE,
    payout_amount DOUBLE
);

CREATE TABLE IF NOT EXISTS Driver_Telemetry_Data (
    Trip_ID STRING,
    policyholder_id BIGINT,
    vin STRING,
    Start_Time STRING,
    End_Time STRING,
    Start_Location_Lat DOUBLE,
    Start_Location_Lon DOUBLE,
    Time_of_Day_Category STRING,
    Duration_Min DOUBLE,
    Distance_Miles DOUBLE,
    Peak_High_Speed DOUBLE,
    Average_Speed DOUBLE,
    Peak_Low_Speed DOUBLE,
    Max_Speed_Limit_Exceeded DOUBLE,
    Sudden_Braking_Count INT,
    Rapid_Acceleration_Count INT,
    Harsh_Cornering_Count INT,
    Speeding_Incidents INT,
    Safety_Score DOUBLE,
    Braking_Index DOUBLE,
    Acceleration_Index DOUBLE,
    Trip_Risk_Level STRING
);
```
