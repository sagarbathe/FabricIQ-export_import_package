# Load AutoClaims Sample Tables

Truncates and reloads all tables from CSV files in the `AutoClaims_csv` folder in the attached lakehouse.

**Prerequisites:**
- Run the **Create AutoClaims Sample Tables** notebook first to create the tables
- Upload the CSV files from the `csv/` folder in this repo to `Files/AutoClaims_csv/` in your lakehouse

```python
table_list = ["accident","adjuster","claim","policy","policyholder","vehicle","driver_telemetry_data"]
```

```python
# Loop through all configured tables and execute the load function
for table in table_list:

    #Get schema of table
    existing_df = spark.read.table(table)
    target_schema = existing_df.schema

    #Read the csv file
    df = spark.read \
        .format("csv") \
        .option("header","true") \
        .schema(target_schema) \
        .load("Files/AutoClaims_csv/"+table+".csv")

    #overwrite into the table
    df.write.mode("overwrite").saveAsTable(table)

    print(f"Table {table} loaded")
```
