# Export Fabric Ontology to Lakehouse

Exports an ontology definition via the Fabric REST API and saves it to a lakehouse Files folder.

**Output files:**
- `<OUTPUT_PATH>/<ONTOLOGY_NAME>_definition.json` — raw API payload (feed this to the Import notebook)
- `<OUTPUT_PATH>/<ONTOLOGY_NAME>_decoded/` — decoded, human-readable JSON parts

**Prerequisites:**
- Create a folder 'FabricIQ-export_import_package' under Files section of your lakehouse. 
- Copy 'fabric_ontology_export-1.0.0-py3-none-any.whl' in that folder
- run the `%pip install` in cell below).

## 2. Configuration

Update the parameters below to match your environment.

**Lakehouse path format:** `abfss://<WorkspaceName>@onelake.dfs.fabric.microsoft.com/<LakehouseName>.Lakehouse/Files`

```python
# ── Source Ontology ──────────────────────────────────────────
WORKSPACE_ID  = "db7dcf85-001e-4277-a85e-3c92029900bc"
ONTOLOGY_ID   = "71280eba-455d-4cad-af82-bef66e09925d"
ONTOLOGY_NAME = "ont_UBI01"

# ── Output Location (lakehouse Files folder) ────────────────
WORKSPACE_NAME  = "WS_AutoClaimsPOC"
LAKEHOUSE_NAME  = "lh_AutoClaims"
EXPORT_FOLDER   = "ontology_exports"

OUTPUT_PATH = (
    f"abfss://{WORKSPACE_NAME}@onelake.dfs.fabric.microsoft.com"
    f"/{LAKEHOUSE_NAME}.Lakehouse/Files/{EXPORT_FOLDER}"
)

print(f"Source:  {ONTOLOGY_NAME} ({ONTOLOGY_ID})")
print(f"Output:  {OUTPUT_PATH}")
```

## 3. Export

```python
from fabric_ontology_export import export_ontology

token = notebookutils.credentials.getToken("https://api.fabric.microsoft.com")

result = export_ontology(
    token=token,
    workspace_id=WORKSPACE_ID,
    ontology_id=ONTOLOGY_ID,
    ontology_name=ONTOLOGY_NAME,
    output_path=OUTPUT_PATH,
)

print(f"\nDefinition file: {result['definition_file']}")
print(f"Decoded folder:  {result['decoded_folder']}")
```
